```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from l0module import L0Linear
```

```python
use_cuda = torch.cuda.is_available()
device = torch.device("cuda" if use_cuda else "cpu")
```

## Data

```python
train_size = 100
test_size = 1000

X_train = torch.from_numpy(np.linspace(1,10,train_size)).to(device)
X_test = torch.from_numpy(np.linspace(1,100,test_size)).to(device)
Y_train = X_train * 3
Y_test = X_test * 3
```

## Model

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(1, 10)
        self.fc2 = nn.Linear(10, 1)
    def forward(self, x):
        x = self.fc1(x)
        x = self.fc2(x)
        return x
    
class SparseNet(nn.Module):
    def __init__(self):
        super(SparseNet, self).__init__()
        self.fc1 = SETLayer(1, 10)
        self.fc2 = L0Linear(10, 1)
    def forward(self, x):
        x = self.fc1(x)
        x, p = self.fc2(x)
        return x, p
```

```python
log_interval = 20

def train(model, optimizer, criterion, epoch):
    model.train()
    result = []
    for i, (x, y) in enumerate(zip(X_train, Y_train)):
        optimizer.zero_grad()
        output = model(x.unsqueeze(0))
        loss = criterion(output, y)
        loss.backward()
        optimizer.step()
#         if i % log_interval == 0:
#             print('[Train] Epoch: {}, Loss: {:.6f}'.format(epoch, loss.item()))
        if model.__class__ == L0Net:
            output = output[0]
        result.append([x.item(), output.item(), y.item()])
    return pd.DataFrame(data=result, columns=['x', 'prediction', 'y'])
        
def test(model, criterion, epoch):
    model.eval()
    test_loss = 0
    result = []
    with torch.no_grad():
        for (x, y) in zip(X_test, Y_test):
            output = model(x.unsqueeze(0))
            test_loss += criterion(output, y).item()
            penalty = 0
            if model.__class__ == L0Net: 
                output, penalty = output[0], output[1]
            result.append([x.item(), output.item(), y.item(), penalty])
        test_loss /= len(X_test)
    print('\n[Test] Epoch: {}, Average Loss: {:.6f}'.format(epoch, test_loss))
    return pd.DataFrame(data=result, columns=['x', 'prediction', 'y', 'sparsity_loss']), test_loss

def run(model, optimizer, criterion, epochs, early_stop=False):
    train_df, test_df, test_loss = None, None, 0
    for epoch in range(epochs):
        train_df = train(model, optimizer, criterion, epoch+1)
        test_df, new_test_loss = test(model, criterion, epoch+1)
        if abs(new_test_loss - test_loss) < 1e-4 and early_stop:
            break
        test_loss = new_test_loss
    return train_df, test_df
```

## Experiments

```python
def plot_df(df):
    '''
    Assumes first column x, other columns y_i
    '''
    xcol = df.columns[0]
    cmap = plt.cm.rainbow(np.linspace(0, 1, len(df.columns[1:])))
    for ycol, color in zip(df.columns[1:], cmap):
        plt.plot(xcol, ycol, data=df, color=color)
        plt.legend()
```

```python
def density(x):
    return torch.norm(x.abs().add(0.9).floor(), 0)
def layer_parameters(layer):
    return torch.cat([x.view(-1) for x in layer.parameters()]) 
```

### Baseline

```python
def criterion(x, y):
    return F.mse_loss(x, y)
model = Net().double().to(device)
optimizer = optim.Adam(params=model.parameters(), lr=1e-3)
train_df, test_df = run(model, optimizer, criterion, 100)
```
